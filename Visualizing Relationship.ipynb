```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the required datasets

```python
df = pd.read_csv('C:/JonathanWorks/ML/Data/Complete_Process_Cars.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(8,8))
```

Finding relationships on the data 

```python
# Relationship between Age and MPG
plt.bar(df['Age'], df['MPG'])

plt.xlabel('Age of Car')
plt.ylabel('Mile Per Gallon')

plt.show()
```

```python
# Relationship between Acceleration and MPG
plt.bar(df['Acceleration'], df['MPG'])

plt.xlabel('Acceleration')
plt.ylabel('Miles Per Gallon')

plt.show()
```

```python
# Relationship between Acceleration and MPG
plt.scatter(df['Acceleration'], df['MPG'])

plt.xlabel('Acceleration')
plt.ylabel('Miles Per Gallon')

plt.show()
```

```python
# Relationship between Weight and MPG
plt.bar(df['Weight'], df['MPG'])

plt.xlabel('Car Weight')
plt.ylabel('Miles per Gallon')

plt.show()
```

```python
# Relationship between Weight and MPG
plt.scatter(df['Weight'], df['MPG'])

plt.xlabel('Car Weight')
plt.ylabel('Miles per Gallon')

plt.show()
```

```python
# Check relation between Cylinders and MPG
plt.scatter(
    x=df['Cylinders'],
    y=df['MPG']
)

plt.xlabel('Cylinder')
plt.ylabel('Miles Per Gallon')

plt.show()
```

```python
# Check relation between Cylinders and MPG
plt.bar(df['Cylinders'],df['MPG'])

plt.xlabel('Cylinder')
plt.ylabel('Miles Per Gallon')

plt.show()
```

```python
df.hist(figsize=(20,10))
plt.show()
```

Dropping the Cylinders and Origin columns 

```python
df.drop(['Cylinders', 'Origin'], axis=1, inplace=True)
```

```python
df.sample(20)
```

Density Estimation Plot

```python
df.plot(
    kind='density',
    subplots=True,
    layout=(3,2),
    sharex=False,
    figsize=(16,24)
)

plt.show() 
```

Box plot

```python
df.plot(
    kind='box',
    subplots=True,
    layout=(3,2),
    sharex=False,
    figsize=(16,24)
)

plt.show()
```

Check correlation 

```python
df_corr = df.corr()
df_corr
```

Correlation in heat map

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.heatmap(df_corr, annot=True)
```
